```python
import pandas as pd
```

```python
dc = pd.read_excel("ikan.xlsx").to_dict()
dc
```

```python
dc1, dc2 = {"Berat ikan (Y)":[], "Luas ukuran Ikan (X)":[]}, {"Berat ikan (Y)":[], "Luas ukuran Ikan (X)":[]}
for spesies, berat, ukuran in zip(list(dc["Spisies Ikan"].values()), list(dc["Berat ikan (Y)"].values()), list(dc["Luas ukuran Ikan (X)"].values())):
    if spesies == "Bream":
        dc1["Berat ikan (Y)"].append(berat)
        dc1["Luas ukuran Ikan (X)"].append(ukuran)
    else:
        dc2["Berat ikan (Y)"].append(berat)
        dc2["Luas ukuran Ikan (X)"].append(ukuran)

df1 = pd.DataFrame(dc1)
df2 = pd.DataFrame(dc2)
```

```python
# Filter noise
dc2 = {"Berat ikan (Y)":[], "Luas ukuran Ikan (X)":[]}
df2 = df2.to_dict()
for x, y in zip(list(df2["Luas ukuran Ikan (X)"].values()), list(df2["Berat ikan (Y)"].values())):
    if (y < 40) or (y > 300): continue
    dc2["Berat ikan (Y)"].append(y)
    dc2["Luas ukuran Ikan (X)"].append(x)
    
df2 = pd.DataFrame(dc2)
df2
```

```python
# hasil setelah filter
plt.scatter(df2["Luas ukuran Ikan (X)"], df2["Berat ikan (Y)"])
plt.xlabel("Luas ukuran Ikan")
plt.ylabel("Berat ikan")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
bream_x = np.array(df1["Luas ukuran Ikan (X)"])
bream_y = np.array(df1["Berat ikan (Y)"])

roach_x = np.array(df2["Luas ukuran Ikan (X)"])
roach_y = np.array(df2["Berat ikan (Y)"])

print(bream_x)
print(bream_y)
print(roach_x)
print(roach_y)
```

```python
bream_x = bream_x.reshape(-1, 1)

roach_x = roach_x.reshape(-1, 1)

print(bream_x)
print(roach_x)
```

```python
# membuat model ML Linear Reggression
model_bream = LinearRegression()
model_roach = LinearRegression()

# panggil method fit untuk melakukan training menggunakan model di atas
model_bream.fit(bream_x, bream_y)
model_roach.fit(roach_x, roach_y)
```

```python
# visualisasi regresi linear bream spesies
X_vis = np.array([40, 130]).reshape(-1, 1)
y_vis = model_bream.predict(X_vis)

plt.scatter(bream_x, bream_y)
plt.plot(X_vis, y_vis, '-r')

plt.title("Visualisasi hasil prediksi Regresi linear ikan Bream")
plt.xlabel("Luas ukuran ikan")
plt.ylabel("Berat ikan")
plt.show()
```

```python
# visualisasi regresi linear roach spesies
X_vis = np.array([5, 50]).reshape(-1, 1)
y_vis = model_roach.predict(X_vis)

plt.scatter(roach_x, roach_y)
plt.plot(X_vis, y_vis, '-r')

plt.title("Visualisasi hasil prediksi Regresi linear ikan Roach")
plt.xlabel("Luas ukuran ikan")
plt.ylabel("Berat ikan")
plt.show()
```

```python
# tampilkan koofesien model regresi ikan spesies Bream
bream_coef = model_bream.coef_
print(f"Koofesien Bream : {bream_coef}")
```

```python
# tampilkan koofesien model regresi ikan spesies Roach
roach_coef = model_roach.coef_
print(f"Koofesien Bream : {roach_coef}")
```
