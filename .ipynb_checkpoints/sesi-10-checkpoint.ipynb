```python
import pandas as pd

'''
Binning
pengelompokan nilai seperti pemberian label pada range data yang di tentukan.
Binning bisa membantu untuk menghindari overfitting. Namun binning juga
mengorbankan informasi yang terkandung dari sebuah atribut sehingga,
penggunaanya perlu dilakukan dengan teliti.
'''

print("======= Binning =======")
df = pd.DataFrame({'Number':[40, 34, 26, 32, 54, 54, 32, 523, 52, 13, 71, 33]})

print("Data Awal :\n", df)
df['bin'] = pd.cut(df['Number'], bins=[0, 30,  70, 100], labels=["Low", "Mid", "High"])
print("\nData Setelah dilakukan Binning :\n", df)

'''
Spliting

Memisahkan sebuah atribut menjadi atribut-atribut baru yang lebih mudah di
pahami oleh model ML kita Contohnya seperti ini.
Misal kita memiliki atribut yang berisi nama seseorang, katakanlah Satria Rahmat
kita dapat membuat atribut baru yang berisi first name dan last name
'''

print("\n\n======= Spliting =======")
df = pd.DataFrame(
    {
        'Name':['Human A', 'Human B', 'Human C'],
        'Age':[40, 34, 26]
    }
)

print("Data Frame Awal\n", df)

df[['First','Last']] = df.Name.str.split(expand=True) 
print("\nData Frame setelah Split kolom 'Name', lalu menambahkan kedua kolom :\n", df)

'''
Interaction Future

Mengkombinasikan dua atribut menjadi atribut yang lebih lebih baik.
Contoh teknik Interaction Future :
- feature selection, teknik lain dalam feature engineering.
- menyusun (handcrafted)
'''
```
